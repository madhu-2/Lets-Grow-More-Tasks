```python
#Loading the all required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a histogram of the number of terrorist attacks by year
plt.hist(df['iyear'], bins=50)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Number of Terrorist Attacks by Year')
plt.show()
```

```python
# Group the data by country and count the number of attacks in each country
country_counts = df.groupby('country_txt')['eventid'].count().sort_values(ascending=False)

# Create a bar chart of the number of attacks by country
plt.figure(figsize=(12,6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Number of Terrorist Attacks by Country')
plt.show()
```

```python
# Group the data by the variable 'column_name' and count the number of occurrences of each category
counts = df.groupby('region')['eventid'].count().sort_values(ascending=False)
# Create the bar chart using the seaborn 'barplot' function
plt.figure(figsize=(12,6))
sns.barplot(x=counts.index, y=counts.values)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Number of Occurrences')
plt.title('Title of the Bar Chart')
plt.show()

```
